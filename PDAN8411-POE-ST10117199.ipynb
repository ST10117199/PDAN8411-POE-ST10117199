#### Text processing 

sourced from:https://medium.com/mlearning-ai/nlp-tokenization-stemming-lemmatization-and-part-of-speech-tagging-9088ac068768

This is the process of taking the text data from its raw data format into a format that is more suitable for analysis in a machine learning model and often used in the data preprocessing format. This is also the most crucial first step needed in turning the raw text data into format that machine learning model can understand. This is a necessary step in text analyse or in natural language processing NLP as it reduces the text data ambiguity and noise making is easier for the models to read the data.
Some of these processes used in NLP include Tokenization, Stemming, Lemmatization, and Part of Speech Tagging

#### Tokenization


 

Tokenization Is the process of processing a given texts or performing natural language processing (NLP) ,by breaking down the sentence into its different parts in smallest unit possible which is referred to as a token . These tokens also include the words and punctuation of the sentence along with other numbers.
This is useful for finding the unique words in the sentences frequency of use using the divided tokens. It is also suitable to be used to tag the token according to the type of the word. These properties of the token ,enable it to be suitable in building models to better analyse the properties of the text data.


#### Stemming 

https://iq.opengenus.org/stemming-in-nlp/
This is the process of assessing a word in order to find the base form of a word or otherwise knows as the words stem. In this example all the words running, runner, and runs will all be treated with the same meaning in essence and changed to base word run as they essentially have the same meaning.Finding roots or stem is done using a rule based process that cuts the suffixes in words according to a certain rule.




There are several advantages of using Stemming that are as follows:
* Increased Accuracy: As this process decreases the quantity of unique word needed to be processed, there by increasing the accuracy of the processing. This results in the quicker and more precise processing. 
* Reduced Vocabulary Size: This is useful with constrained memory and processing requirement by reducing the vocabulary size necessary  for NLP 
* Easier Preprocessing: This aid in analysing the assessing various instances of the same word by condensing it into a single word .This make preprocessing, cleaning and normalizing text simpler.
Some disadvantages include:
* Loss of Details- This can make some crucial details of a words part of speech or tense  loose vital details making less accurate or misleading outcomes.
This also as 2 other problems that can occur should stemming not be implemented properly 

These problems include overstemming and understemming.
* Overstemming- This occurs with excessive letters being removed leading to stems not actually a word. This can cause outcomes errors and problem processing.
* Understemming- This occurs with insufficient prefixes or suffixes being removed leading to stems not actually in their base for or base stem. This can also cause outcomes errors and problem processing.

#### Lemmatization

https://www.engati.com/glossary/lemmatization#:~:text=Lemmatization%20is%20extremely%20important%20because,much%20slower%20than%20stemming%20algorithms.

Lemmatization is the process of assessing the word and comparing it to a related word in the dictionary. This also has a similar purpose to stemming as is used to reduce inflectional forms to a common base form . Unlike stemming that removes inflations , Lemmatization makes use of lexical knowledge bases in order to get a suitable base form of the word.




There are several advantages of lemmatization: 
* lemmatization is more accurate then Stemming. Its suitable to get the root words from the dictionary as apposed to stemming that cuts the data without context.
* Lemmatization is able to give context on the exact and contextual meaning. 
There are also several disadvantages of lemmatization:
* This is a slow and time consuming process, as it asseses using the dictionary context of the words
* Lemmatization is much slower then stemming.


Part of Speech Tagging




Part of Speech Tagging is a process of using the word types like noun ,verbs , adverbs, adjectives and others to label the word in the text and converts a sentence to forms consisting of the words list and  tuples or tag list .This assesses the feature of previous word , next word ,capitalised first letter , along with others in supervised learning solution.

#### What are stop words 

This is a set of English words that are commonly used in language sentances .These common words do not however add much meaning to the sentance ,and can be ignored without changing the meaning of the sentance. The fuction stopwords is a collection of these words that can be downloaded with some imports to the python enviroment. Some common stop words include 'a','the','is','are', etc...

https://kavita-ganesan.com/what-are-stop-words/

```python
# Assigning additional words to stopwords from the stop words  
SW = stopwords.words('english') + ['u', '', 'ur', '4', '2', 'im', 'dont', 'doin', 'ure']
str(SW)
```

#### Stemming

```python
# Note below is sourced from Class work SpamHamText
# Removes all punctuation, stopwords and returns a list of the cleaned text
def Stemming(Inputwords): 
    # Compares each item from the inputed words to each item in the imported list of all punctuation and removes the ones that match 
    PunctuationRemoved = [item1 for item1 in Inputwords if item1 not in string.punctuation]
    PunctuationRemoved = ''.join(PunctuationRemoved)
    # Compares each item from the inputed that have had thier punctuation fixed to each item in the imported list of Stopwords to remove the ones that match 
    stopwordsRemoved = [item2 for item2 in PunctuationRemoved.split() if item2.lower() not in SW]
    stopwordsRemoved = ' '.join(stopwordsRemoved)
    return stopwordsRemoved
```

```python
# Calling the method above and asigning the result in new column
df['Stem_title'] = df.title.apply(Stemming)
df['Stem_text'] = df.text.apply(Stemming)
df.head(2)
```

### Lemmatization with tokenizer

```python
tokenizers = nltk.tokenize.WhitespaceTokenizer()
lemmatizers = nltk.stem.WordNetLemmatizer()
```

```python
# Making a function ued for peformining WordLemmatization and returns a list of the cleaned text
def WordLemmatization(Inputwords): 
	# splits the sentence entered into parts 
	h=tokenizers.tokenize(Inputwords)
	k=''
	# loops through each part and applies lemmatization
	for w in h:
			k=k+" "+lemmatizers.lemmatize(w)
	return k
```

```python

df['lem_title'] = df.Stem_title.apply(WordLemmatization)
df['lem_text'] = df.Stem_text.apply(WordLemmatization)
df.head(2)
```

```python
def testTrueAndFake(testsize):
    #pefroming train test split
    x_train, x_test, y_train, y_test = train_test_split(X_U_TorF, Y_U_TorF, test_size =testsize, random_state = 555)
    # calling the SKlearn function for a Naive Bayes Model
    CV.fit(x_trainTF)
    XTRAIN = CV.fit_transform(x_train)
    XTEST = CV.transform(x_test)
    # Below is used to aply the test data against the training data 
    MLNB.fit(XTRAIN, y_train)
    #below is used to make a prediction on the test data 
    pred = MLNB.predict(XTEST)
    # Below is used to get the accurcy score and return a range of answers.
    MLNBAcc=accuracy_score(y_test,pred)
    # storing data to be returned in a list format.
    disp=[[testsize,MLNBAcc]]
    # returns the stored values.
    return disp
```

```python
# Define the initial testsize to 0
testsize = 0
# Define a empty list
answer=[[]]
# Create a while loop that will loop till the test size <0.95
while testsize < 0.95:
    #increments the tests size till the test size reaches 0.95 
    testsize = testsize+ 0.05
    #adds all the results of each icrement in test size to a list 
    answer = answer+testTrueAndFake(testsize)
#converts the list to a dataframe
KNN_Acc= pd.DataFrame(answer)
#This just removes the first null value as this fromat creates null value by default that is not needed.
KNN_Acc = KNN_Acc.iloc[1:]
# Rename the columns to somthing more appropriate.
KNN_Acc.rename(columns=
    {
    0:"test size",
    1:"Accuracy"
    }
    ,inplace=True)
# Display results
KNN_Acc.head().sort_values('Accuracy', ascending=False)
```
